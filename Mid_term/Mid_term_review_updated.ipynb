# for-loop

```python
# given 2 lists as
ls1 = list(range(10))
ls2 = [i%2==0 for i in ls1]
```

```python
#zip ls1 and ls2 together like  "Is 1 an even number? False", "Is 2 an even number? True"
```

```python
for i in range(len(ls1)):
    print(f"Is {ls1[i]} an even number? {ls2[i]}")    
```

```python
#use an empty list to collect 
collection_bucket = []
for i in range(len(ls1)):
    collection_bucket.append(f"Is {ls1[i]} an even number? {ls2[i]}")
    
collection_bucket
```

```python
#use an empty dict to collect 
collection_bucket = {}
for i in range(len(ls1)):
    collection_bucket[i] = f"Is {ls1[i]} an even number? {ls2[i]}"
    
collection_bucket
```

*** The best self-defined function is robust on exhaustive conditions ***

```python
#good practice is to limit the input adn output type

def sumList(a : list, b: list) -> list:
    
    try:
        '''check the list should have same lenght'''
        #assert len(a) == len(b), "the lenght of a and b are different"
    
        out = []
        for i in range(len(a)):
            out.append(a[i] + b[i])   
            return out
        return out               
    except:
        print("the lenght of a and b are different")

    
```
